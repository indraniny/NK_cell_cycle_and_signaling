```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import fcsextract
from scipy import stats


#greg's gating 
path='./data/primary/greg_umaps/untransformed/debarcode_IL2_NKG2D_'
filename_1=path+'2min_CD56dim_UMAP_G1.fcs'
filename_2=path+'2min_CD56dim_UMAP_S.fcs'

#plotting histogram of the protein

 # Use fcsextract to parse the FCS file
metadata_1, data_1 = fcsextract.fcsextract(filename_1)

metadata_2, data_2 = fcsextract.fcsextract(filename_2)


param_names_1 = []
for i in range(int(metadata_1[b'$PAR'])):
      param_names_1.append(metadata_1.get(b"$P%dS"%(i+1),metadata_1[b"$P%dN"%(i+1)]))
data_1 = np.asarray(data_1)

param_names_2 = []
for i in range(int(metadata_1[b'$PAR'])):
      param_names_2.append(metadata_2.get(b"$P%dS"%(i+1),metadata_2[b"$P%dN"%(i+1)]))
data_2 = np.asarray(data_2)

#index=22
#------------------------------------------------
index = param_names_1.index(b'CD7')
print(index)

print(param_names_1[index])
print(param_names_2[index])



Y1=data_1[:,index]
Y2=data_2[:,index]

#--------------------------------------------------------
#data processing
Y1[Y1<0]=1e-11
Y2[Y2<0]=1e-11
Y1=np.log(Y1)
Y2=np.log(Y2)

#---------------------------------------------------------
mean_1=np.mean(Y1)
mean_2=np.mean(Y2)

std_1=np.std(Y1)
std_2=np.std(Y2)

print(f'data1={Y1}')
print(f'data2={Y2}')


print(np.shape(Y1),np.shape(Y2))

# Perform two-sample t-test
t_stat, p_value = stats.ttest_ind(Y1, Y2, equal_var=False,alternative='less') 


print(f' mean1= {mean_1} mean2={mean_2}')
print(f' std_1= {std_1} std2={std_2}')

print("\n")

print("t-statistic:", t_stat)
print("p-value:", p_value)

print("\n")

if (mean_2 > mean_1):
      print( 'increase in protein') 
elif (mean_2 < mean_1):
      print( 'decrease in protein') 
elif (mean_1==mean_2):     
      print ( ' no change')

if ( p_value < 0.05):
      print('with significance') 
else:
      print ('no significance')
print(param_names_1)
```
