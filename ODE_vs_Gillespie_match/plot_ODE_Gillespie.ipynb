```python
import matplotlib.pyplot as plt
import scipy.io


# Create the figure
#plt.figure(figsize=(10, 6))

# Load data from .mat files
times_data = scipy.io.loadmat('times.mat')  # Load times data
Gillespie_protein_G1_data = scipy.io.loadmat('Gillespie_protein_G1_levels.mat')
Gillespie_protein_S_data = scipy.io.loadmat('Gillespie_protein_S_levels.mat')
Gillespie_protein_G2_data = scipy.io.loadmat('Gillespie_protein_G2_levels.mat')
Gillespie_protein_M_data = scipy.io.loadmat('Gillespie_protein_M_levels.mat')
A_levels_ode_data = scipy.io.loadmat('A_levels_ode.mat')


# # Extract the actual data arrays from the loaded dictionaries
times = times_data['times'].flatten()  # Adjust the key name if necessary
print(times)
print(type(times))

Gillespie_protein_G1_levels = Gillespie_protein_G1_data['protein_G1_levels']
#print(type(Gillespie_protein_G1_levels))

Gillespie_protein_S_levels = Gillespie_protein_S_data['protein_S_levels']
Gillespie_protein_G2_levels = Gillespie_protein_G2_data['protein_G2_levels']
Gillespie_protein_M_levels = Gillespie_protein_M_data['protein_M_levels']
A_levels_ode = A_levels_ode_data['A_levels_ode']

# Set global font settings
plt.rcParams.update({
    'font.family': 'Arial',  # Use Arial font
    'font.size': 16,         # Set font size
    'axes.labelsize': 16,    # Size of axis labels
    'xtick.labelsize': 16,   # Size of x-tick labels
    'ytick.labelsize': 16,   # Size of y-tick labels
})

colors = ['lightblue', '#069AF3', 'orange', '#EC5800']

# # Plot protein levels for G1, S, G2, M
# Plot protein levels for G1, S, G2, M with different colors

# ODE plots with markers
plt.plot(times, A_levels_ode[0, :], color=colors[0], linewidth=2, label='G1 (ODE)')
plt.plot(times, A_levels_ode[1, :], color=colors[1], linewidth=2, label='S (ODE)')
plt.plot(times, A_levels_ode[2, :], color=colors[2], linewidth=2,  label='G2 (ODE)')
plt.plot(times, A_levels_ode[3, :], color=colors[3], linewidth=2,  label='M (ODE)')


# Gillespie plots
plt.plot(times, Gillespie_protein_G1_levels[:,0], label='G1 (KME)', color=colors[0], linestyle=':',linewidth=2)
plt.plot(times, Gillespie_protein_S_levels[:,0],label='S (KME)', color=colors[1], linestyle=':',linewidth=2)
plt.plot(times, Gillespie_protein_G2_levels[:,0], label='G2 (KME)', color=colors[2],linestyle=':', linewidth=2)
plt.plot(times, Gillespie_protein_M_levels[:,0], label='M (KME)', color=colors[3], linestyle=':',linewidth=2)



# Title and labels
# plt.title('Gillespie: Protein Levels Across Cell Cycle Stages')
# plt.xlabel('Time')
# plt.ylabel('Avg. Protein Level')


# Customize the font of the legend
plt.legend(fontsize=14, prop={'family': 'arial'})



# Set the width of the x and y axes (spines)
ax = plt.gca()
axes_width_size=2
ax.spines['bottom'].set_linewidth(axes_width_size)
ax.spines['left'].set_linewidth(axes_width_size)
ax.spines['top'].set_linewidth(axes_width_size)  # Optional: set top spine width to 0 if not needed
ax.spines['right'].set_linewidth(axes_width_size) 



# Add legend
plt.legend()

# Show the plot
plt.show()
```
